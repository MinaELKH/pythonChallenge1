```python
##  question 8 :  Pour chaque commande, afficher les plats commandés, leur quantité, et le coût total des ingrédients.
# select c.id , p.nom , cp.quantite , sum(cp.quantite *pi.quantite_necessaire * i.cout_unitaire) as cout_Total_ingredients from commandes c
# join commande_plats cp on c.id = cp.commande_id
# join plats p on p.id = cp.plat_id
# join plat_ingredients pi on  p.id = pi.plat_id
# join ingredients i on  i.id = pi.ingredient_id
# group by c.id , p.nom , cp.quantite 
# order by c.id

from sqlalchemy import select , func

cmd = commandes.alias("cmd")
p = plats.alias("p")
cp = commande_plats.alias("cp")
pi = plat_ingredients.alias("pi")
i = ingredients.alias("i") 

stmt = (
    select(
        cmd.c.id.label("id_cmd"),
        p.c.nom.label("plat"),
        cp.c.quantite.label("quantite"),
        func.sum(cp.c.quantite * pi.c.quantite_necessaire * i.c.cout_unitaire)
            .label("cout_total_ingredients")
    )
    .join(cp, cmd.c.id == cp.c.commande_id)
    .join(p, p.c.id == cp.c.plat_id)
    .join(pi, p.c.id == pi.c.plat_id)
    .join(i, i.c.id == pi.c.ingredient_id)
    .group_by(cmd.c.id, p.c.nom, cp.c.quantite)
    .order_by(cmd.c.id)
)


with engine.connect() as conn:
    rslt = conn.execute(stmt)
    for record in rslt:
        print("\n", record)
```

```python
##  question 9 :  Afficher le nombre de plats pour chaque catégorie, y compris celles sans plats.

from sqlalchemy import select, func

catg = categories.alias("cat")
p = plats.alias("p")

stmt = (
    select(
        catg.c.nom.label("nom_categorie"),
        func.count(p.c.id).label("nombre_de_plats")
    )
    .select_from(catg.outerjoin(p, catg.c.id == p.c.categorie_id))  
    .order_by(catg.c.nom)
)

with engine.connect() as conn:
    rslt = conn.execute(stmt)
    for record in rslt:
        print(record)
```

```python
##  question 10  : Afficher le prix moyen des plats par catégorie et le coût moyen des ingrédients par plat.
from sqlalchemy import select, func

catg = categories.alias("cat")
p = plats.alias("p")
cp = commande_plats.alias("cp")
pi = plat_ingredients.alias("pi")
i = ingredients.alias("i")  


stmt = (
    select(
        catg.c.nom.label("nom_categorie"),
        func.avg(p.c.prix).label("moyen_prix"),
        func.avg(pi.c.quantite_necessaire * i.c.cout_unitaire).label("cout moyen ingredien")
    )
    .join(p, catg.c.id == p.c.categorie_id)
    .join(pi, p.c.id == pi.c.plat_id)
    .join(i, i.c.id == pi.c.ingredient_id)
    .group_by(catg.c.nom)
    .order_by(catg.c.nom)
)

with engine.connect() as conn:
    rslt = conn.execute(stmt)
    for record in rslt:
        print(record)
```

```python
# question 13 : Lister les plats commandés plus de trois fois avec leur total de quantités et leur note moyenne (via avis).
from sqlalchemy import select, func

plt = plats.alias("plt")
cp = commande_plats.alias("cp")
cmd = commandes.alias("cmd")
av = avis.alias("av")

stmt = (
    select(
        plt.c.nom.label("plat"),
        func.sum(cp.c.quantite).label("total_quantite"),
        func.avg(av.c.note).label("note_moyenne")
    )
    .select_from(cp)
    .join(cmd, cmd.c.id == cp.c.commande_id)
    .join(plt, plt.c.id == cp.c.plat_id)
    .outerjoin(av, av.c.plat_id == plt.c.id) 
    .group_by(plt.c.nom)
    .having(func.sum(cp.c.quantite) >= 3)   
    .order_by(func.sum(cp.c.quantite).desc())
)

with engine.connect() as conn:
    rslt = conn.execute(stmt)
    for record in rslt:
        print(record)
```

```python

# question 20  : Afficher pour chaque client :
    # Son nom
    # Le nombre total de plats commandés
    # Le montant total dépensé
    # La note moyenne de leurs avis

from sqlalchemy import select, func

plt = plats.alias("plt")
cp = commande_plats.alias("cp")
cmd = commandes.alias("cmd")
av = avis.alias("av")
clt = clients.alias("clt")
stmt = (
    select(
        clt.c.nom.label("nomClient") , 
        func.sum(cp.c.quantite).label("total_quantite"),
        func.sum(cmd.c.total).label("total_depense"),
        func.avg(av.c.note).label("note_moyenne")
    )
    .select_from(clt)
    .join(cmd , clt.c.id == cmd.c.client_id)
    .join(cp, cmd.c.id == cp.c.commande_id)
    .join(plt, plt.c.id == cp.c.plat_id)
    .outerjoin(av, av.c.plat_id == plt.c.id) 
    .group_by(clt.c.nom)
    .order_by(func.sum(cp.c.quantite).desc())
)

with engine.connect() as conn:
    rslt = conn.execute(stmt)
    for record in rslt:
        print(record)
```

```python
#  question 22 : Afficher les clients et leurs dernières commandes, incluant les plats commandés.

from sqlalchemy import select , func , desc

cmd = commandes.alias("cmd")
plt = plats.alias("plt")
cp = commande_plats.alias("cp")
clt = clients.alias("clt")


stmt = (
    select(
        clt.c.id.label("client_id"), 
        func.max(cmd.c.date_commande).label("derniere_date")
    )
    .select_from(clt)
    .join(cmd , cmd.c.client_id == clt.c.id)
    .group_by(clt.c.id)
    .subquery()
)

stmt1 = (
    select(
        clt.c.nom.label("nom_client"),
        cmd.c.id.label("id_cmd"),
        cmd.c.date_commande.label("date_cmd"),
        plt.c.nom.label("plat")
    )
    .select_from(clt)
    .join(cmd, cmd.c.client_id == clt.c.id)
    .join(stmt, (stmt.c.client_id == clt.c.id) & (stmt.c.derniere_date == cmd.c.date_commande))
    .join(cp, cp.c.commande_id == cmd.c.id)
    .join(plt, plt.c.id == cp.c.plat_id)
    .order_by(clt.c.nom)
)

with engine.connect() as conn:
    rslt = conn.execute(stmt1)
    for record in rslt:
        print(record)
```

```python
# question 23 : # Créer une vue virtuelle (SELECT) qui affiche :
            # Le nom du client
            # Les plats commandés
            # Les quantités
            # La date de la commande
            # La note moyenne du plat (via avis)

from sqlalchemy import select, func

plt = plats.alias("plt")
cp = commande_plats.alias("cp")
cmd = commandes.alias("cmd")
av = avis.alias("av")
clt = clients.alias("clt")

stmt = (
    select(
        clt.c.nom.label("nomClient"),
        plt.c.nom.label("nomplat"),
        cp.c.quantite.label("quantite"),
        cmd.c.date_commande.label("date_Commande"),
        func.avg(av.c.note).label("note_moyenne")
    )
    .select_from(clt)
    .join(cmd, clt.c.id == cmd.c.client_id)
    .join(cp, cmd.c.id == cp.c.commande_id)
    .join(plt, plt.c.id == cp.c.plat_id)
    .outerjoin(av, av.c.plat_id == plt.c.id)
    .group_by(clt.c.nom, plt.c.nom, cp.c.quantite, cmd.c.date_commande)  
    .order_by(clt.c.nom, cmd.c.date_commande.desc())
)


##  cette methode permet de transmettre la requete alchemy , to requete sql puis via text il execute sous forme de view 
compiled_sql = stmt.compile(engine, compile_kwargs={"literal_binds": True})


with engine.begin() as conn:
    conn.execute(
        text(f"CREATE OR REPLACE VIEW vue_client_commandes AS {compiled_sql}")
    )
```
